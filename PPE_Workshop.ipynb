**Visualizing YOLO bounding box annotations on the image.**

```python
import cv2
import matplotlib.pyplot as plt

# file paths
img_path = "/content/image.png"
label_path = "/content/image.txt"

# class names and colors
classes = {
    0: "Person",
    1: "Face-mask-medical",
    2: "Gloves",
    3: "Helmet",
    4: "Safety-suit",
    5: "Safety-vest",
}
colors = [
    (255, 0, 0), (0, 255, 0), (0, 0, 255),
    (255, 255, 0), (255, 0, 255), (0, 255, 255)
]

# load image
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w = img.shape[:2]

# load labels
labels = []
for line in open(label_path):
    c, xc, yc, bw, bh = map(float, line.split())
    labels.append((int(c), xc, yc, bw, bh))

# draw boxes
for c, xc, yc, bw, bh in labels:
    x1 = int((xc - bw/2) * w)
    y1 = int((yc - bh/2) * h)
    x2 = int((xc + bw/2) * w)
    y2 = int((yc + bh/2) * h)
    color = colors[c % len(colors)]
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img, classes.get(c, str(c)), (x1, y1 - 5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

# show result
plt.imshow(img)
plt.axis("off")
plt.show()
```

**Building random 10% train and validation subsets to accelerate model training.**

```python
import os, shutil, random

base = "/content/filtered"
images_dir = f"{base}/images"
labels_dir = f"{base}/labels"

def make_subset(list_file, out_name, pct=0.1):
    os.makedirs(f"{images_dir}/{out_name}", exist_ok=True)
    os.makedirs(f"{labels_dir}/{out_name}", exist_ok=True)

    # Read paths like: images/pexels-photo.jpeg
    with open(list_file) as f:
        files = [line.strip().split("/")[-1] for line in f]  # extract filename only

    random.shuffle(files)
    subset_size = max(1, int(len(files) * pct))
    subset = files[:subset_size]

    for fname in subset:
        shutil.copy(f"{images_dir}/{fname}", f"{images_dir}/{out_name}/")
        shutil.copy(f"{labels_dir}/{fname.rsplit('.',1)[0]}.txt", f"{labels_dir}/{out_name}/")

# Create 10% subsets
make_subset(f"{base}/train.txt", "train_small")
make_subset(f"{base}/val.txt", "val_small")
```
